```python
from tabulate import tabulate
lista_principal = []
#se pide ingresar los datos
m = float(input("m: "))
xo = float(input("xo: "))
a = float(input("a: "))
c = float(input("c: "))
x=0
#Ciclo para operar los datos
for x in range(30):
  xnn=(a*xo)+c
  mod=xnn%m
  division=xo/m
  lista_principal.append([x, xo, xnn, mod, division])
  xo=mod
headers = ["x", "xo", "xnn", "mod","división"]
print(tabulate(lista_principal, headers=headers, tablefmt="grid"))
```
