```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
#n=10
#lam=2
#l= -1/lam
```

```python
def expo(lam=2,n=10):
    l= -1/lam
    #lam,n= 2,10
    list=[]
    for i in range(n)
        u=rnd.random()
        # x=(l*np.log(u))
        if u>=0.5:
          x=1
        else:
          x=0
        list.append(x)
    return list
```

```python
#usa el algoritmo
g= expo(.5,10)
print(g)
```

```python

#calcula media
mu=np.mean(g)
print(mu)
```

```python

#calcula varianza
sigma=np.var(g)
print(sigma)
```

```python
def cdf(g):
  n=len(g)
  x=np.sort(g)
  y=np.arange(0,n)/n
  return x,y
```

```python
x,y=cdf(g)
```

```python

plt.step(x,y)
#media
plt.axvline(mu,color="green")
#mediana
plt.axhline(.5,color="red")
```
