```python
from sympy import symbols, integrate

#Comprobamos que la integral da 1:

#Definimos la variable
x = symbols('x')

#Definimos las funciones en los intervalos
f1 = -1/2 * (x - 4) + 3/4
f2 = 1/2 * (x - 5) + 1/4

#Calculamos las integrales definidas
integral1 = integrate(f1, (x, 4, 5))
integral2 = integrate(f2, (x, 5, 6))

#Sumamos las integrales
total_integral = integral1 + integral2

total_integral.evalf()
```

Usamos el método de la transformada inversa:

f(x) = \begin{cases}
-\frac{1}{2} (x - 4) + \frac{3}{4} & \text{para } 4 \leq x < 5 \\
\frac{1}{2} (x - 5) + \frac{1}{4} & \text{para } 5 \leq x \leq 6
\end{cases}

$
F(x) = \int_4^x \left( -\frac{1}{2}(t - 4) + \frac{3}{4} \right) dt
$


$
F(x) = F(5) + \int_5^x \left( \frac{1}{2}(t - 5) + \frac{1}{4} \right) dt
$

Invertiremos \( F(x) \) para obtener la IDF.

```python
def funcion(N=50):
  Lista = []
  for i in range(N):
    u = rd.uniform(0,1)
    y = rd.uniform(4,6)
    if y <= 5:
      f = (11 - 2*y)/4
    else:
      f = (2*y -9)/4
    if u <= f:
      x = y
      Lista.append(x)
  return Lista

F = funcion(500000)

plt.hist(F,bins = 100)   #mostramos el histograma
plt.show()
```
